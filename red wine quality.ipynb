### EDA analysis

```python
# wine with quality 5,6 generally has more outliers
# wine with high amounts of citric acid and alcohol tend to be high quality
# the 25% and 50% for chlorides and fixed acidity variance is low which could explain the amount of outlier
```

```python
# make new column called 'rating' to categorize quality as bad, average, and good
# quality 3, 4 are bad and have the value 0
# quality 5, 6 are average and have the value 1
# quality 7, 8 are good and have the value 2
rating =[]   
for item in data['quality']:
    if item <= 4:
        rating.append(0)
    elif item <= 6:
        rating.append(1)
    else:
        rating.append(2)
            
```

```python
# add new rating column to dataset
data['rating'] = pd.Series(rating)
```

```python
data.columns
```

```python
# plot number of rating for each category
sns.countplot(x='rating', data=data)
```

### Prediction and model include outliers

```python
X = data[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
           'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density','pH', 'sulphates', 'alcohol']]
y = data['rating']
```

```python
# scale the dataset
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X = scale.fit_transform(X)
```

```python
# PCA the data
from sklearn.decomposition import PCA
pca = PCA()
X_pca = pca.fit_transform(X)
```

```python
X_pca.shape
```

```python
# split data to train and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size = 0.3, random_state=42)
```

```python
# Logistic Regression model
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
logreg_predict = logreg.predict(X_test)
```

```python
# calculat Logistic Regression model accuracy
from sklearn.metrics import confusion_matrix
logreg_score = logreg.score(X_test, y_test)*100
print(confusion_matrix(y_test, logreg_predict))
print(logreg_score)
```

```python
# KNeighbors Classifier model
from sklearn.neighbors import KNeighborsClassifier
knc = KNeighborsClassifier(n_neighbors =3)
knc.fit(X_train, y_train)
knc_predict = knc.predict(X_test)
```

```python
# calculate KNeighbors Classifier model accuracy
knc_score = knc.score(X_test, y_test)*100
print(confusion_matrix(y_test, knc_predict))
print(knc_score)
```

```python
# Support Vector Machine model
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, y_train)
svc_predict = svc.predict(X_test)
```

```python
# calculate Support Vector Machine model accuracy
svc_score = svc.score(X_test, y_test)*100
print(confusion_matrix(y_test, svc_predict))
print(svc_score)
```

```python
# Decision Tree Classifier model
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
dtc_predict = dtc.predict(X_test)
```

```python
# calculate Decision Tree Classifier model accuracy
dtc_score = dtc.score(X_test, y_test)*100
print(confusion_matrix(y_test, dtc_predict))
print(dtc_score)
```

```python
# Random Forest Classifier model
from sklearn.ensemble import RandomForestClassifier
rtc = RandomForestClassifier(n_estimators=100)
rtc.fit(X_train, y_train)
rtc_predict = rtc.predict(X_test)
```

```python
# calculate Random Forest Classifier model accuracy
rtc_score = rtc.score(X_test, y_test)*100
print(confusion_matrix(y_test, rtc_predict))
print(rtc_score)
```

```python
# dataframe of the best prediction score
models = pd.DataFrame({'model': ['Logistic Regression', 'KNeighbors Classifier', 'Support Vector Machine',
                                'Decision Tree Classifier', 'Random Tree Classifier'],
                       'score': [logreg_score, knc_score, svc_score, dtc_score, rtc_score]})

models.sort_values(by='score', ascending=False)
```

```python
# Random Tree Classifier best model for predicting wine quality
```

### Prediction and model exclude outliers

```python
# exclude outliers with low variance in 25% and 50%
x = data[['fixed acidity', 'volatile acidity', 'citric acid','free sulfur dioxide', 'total sulfur dioxide', 'density','pH', 'alcohol']]
Y = data['rating']
```

```python
# scale x, fit and transform
x_scale = StandardScaler()
x = x_scale.fit_transform(x)

# PCA x, fit and transform 
pca_x = PCA(n_components=8)
x_pca = pca_x.fit_transform(x)
```

```python
# split data to train and test set
x_train, x_test, Y_train, Y_test = train_test_split(x_pca, Y, test_size=0.3, random_state = 42)
```

```python
# Logistic Regression model
lr_model = LogisticRegression()
lr_model.fit(x_train, Y_train)
lr_predict = lr_model.predict(x_test)

# model accuracy
lr_score = lr_model.score(x_test, Y_test)*100
print(confusion_matrix(Y_test, lr_predict))
print(lr_score)
```

```python
# KNeighobors Classifier model
kc_model = KNeighborsClassifier(n_neighbors=3)
kc_model.fit(x_train, Y_train)
kc_predict = kc_model.predict(x_test)

# model accuracy
kc_score = kc_model.score(x_test, Y_test)*100
print(confusion_matrix(Y_test, kc_predict))
print(kc_score)
```

```python
# Support Vector Machine
sc_model = SVC()
sc_model.fit(x_train, Y_train)
sc_predict = sc_model.predict(x_test)

# model accuracy
sc_score = sc_model.score(x_test, Y_test)*100
print(confusion_matrix(Y_test, sc_predict))
print(sc_score)
```

```python
# Decision Tree Classifier
dc_model = DecisionTreeClassifier()
dc_model.fit(x_train, Y_train)
dc_predict = dc_model.predict(x_test)

# model accuracy
dc_score = dc_model.score(x_test, Y_test)*100
print(confusion_matrix(Y_test, dc_predict))
print(dc_score)
```

```python
# Random Tree Classifier
rc_model = RandomForestClassifier(n_estimators=100)
rc_model.fit(x_train, Y_train)
rc_predict  = rc_model.predict(x_test)

# model accuracy
rc_score = rc_model.score(x_test, Y_test)*100
print(confusion_matrix(Y_test, rc_predict))
print(rc_score)
```

```python
# dataframe of the best prediction score exclude outliers
xo_model = pd.DataFrame({'model': ['Logistic Regression', 'KNeighbors Classifier', 'Support Vector Machine',
                                'Decision Tree Classifier', 'Random Tree Classifier'],
                       'score': [lr_score, kc_score, sc_score, dc_score, rc_score]})

xo_model.sort_values(by='score', ascending=False)
```
